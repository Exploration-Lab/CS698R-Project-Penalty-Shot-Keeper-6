```python
#Import the necessary libraries for plotting
import time
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

# Import the primary function from Smurves
from smurves import surgebinder

# Generate curves with the parameters requested
curves = surgebinder(n_curves = 100,
                     x_interval = [0.0, 90.0],
                     y_interval = [0.0, 2.0],
                     n_measure = 91,
                     direction_maximum = 50,
                     convergence_point = [0.0, 1.0])

# Set the figure size
f, ax = plt.subplots(figsize=(10, 5))

# Plot each curve in the list of curves
for i in range(0, len(curves)):
    plt.plot(np.asarray(curves[i])[:, 0], np.asarray(curves[i])[:, 1])

# Specify the x-axis and y-axis intervals
x_interval = [0.0, 90.0]
y_interval = [0.0, 2.0]

# Limit the plots to the above intervals
plt.ylim(y_interval[0], y_interval[1])
plt.xlim(x_interval[0], x_interval[1])
```
